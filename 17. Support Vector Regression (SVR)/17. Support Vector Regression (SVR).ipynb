6.ใช้ displot เพื่อดูการกระจายของแต่ละคอลัมน์


```python
columns = list(df.describe().T.index)
fig, axs = plt.subplots(nrows=13, figsize=(30, 30))
for i in range(13):
    sns.distplot(df[columns[i]], ax=axs[i])
    plt.title(columns[i])
```

7.ใช้ heatmap ดูความสัมพันธ์ของคอลัมน์ที่สนใจ


```python
fig = plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), annot=True);
```
