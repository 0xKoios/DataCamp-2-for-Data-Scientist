5. เช็คว่ารีวิวมีปีไหนบ้าง ละแต่ละปีมีจำนวนรีวิวเท่าไหร่


```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.groupby("year").aggregate("count")[["review_id"]]
```

```python
df.groupby("year").aggregate("count")[["review_id"]].plot.bar(figsize=(10, 6));
```

6. เช็คว่ารีวิวมีเดือนไหนบ้าง ละแต่ละเดือนจำนวนรีวิวเท่าไหร่




```python
df.groupby("month").aggregate("count")[["review_id"]]
```

```python
df.groupby("month").aggregate("count")[["review_id"]].plot.bar(figsize=(10, 6));
```

7. เช็คเดือนและปีที่เท่าไหร่มีคนรีวิวมากที่สุด และน้อยที่สุด


```python
df_ = df.groupby("year").aggregate("count")[["review_id"]]
df_[df_["review_id"] == df_["review_id"].max()]
```

```python
df_ = df.groupby("month").aggregate("count")[["review_id"]]
df_[df_["review_id"] == df_["review_id"].max()]
```

```python
df_ = df.groupby("year").aggregate("count")[["review_id"]]
df_[df_["review_id"] == df_["review_id"].min()]
```

```python
df_ = df.groupby("month").aggregate("count")[["review_id"]]
df_[df_["review_id"] == df_["review_id"].min()]
```
