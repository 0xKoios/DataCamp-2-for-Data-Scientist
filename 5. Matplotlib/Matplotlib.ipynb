```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 

```python
x = np.arange(100)
y = np.power(x, 2)
```

```python
plt.figure(figsize=[10, 6])
plt.title("Exponential", size=20)
plt.xlabel("x", size=20)
plt.ylabel("y = x^2", size=20)
plt.plot(x, y, linewidth=6);
```

## 2. สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู 

```python
x = np.arange(180)/(2*np.pi)
y = np.sin(x)
```

```python
plt.figure(figsize=[20, 6])
plt.title("Function sin(x)", size=20)
plt.xlabel("x", size=20)
plt.ylabel("sin(x)", size=20)
plt.plot(x, y, color="pink", linewidth=6);
```

## 3. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก

```python
y = np.arange(100)
x = np.power(y, 2)
```

```python
plt.figure(figsize=[10, 6])
plt.title("Exponential", size=20)
plt.xlabel("y", size=20)
plt.ylabel("x = y^2", size=20)
plt.plot(x, y, linewidth=6);
```

## 4. สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - - 

```python
x = np.arange(100)/(2*np.pi)
y = np.cos(x)
```

```python
plt.figure(figsize=[20, 6])
plt.title("Function sin(x)", size=20)
plt.xlabel("x", size=20)
plt.ylabel("cos(x)", size=20)
plt.plot(x, y, color="red", linewidth=6, linestyle="--");
```

## 5. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker

```python
x = np.arange(0, 100, 5)
y_1 = np.power(x, 2)
y_2 = np.power(x, 3)
```

```python
plt.figure(figsize=[20, 9])

plt.subplot(1, 2, 1)
plt.title("Y = X^2", size=20)
plt.xlabel("X", size=20)
plt.ylabel("Y", size=20)
plt.plot(x, y_1, color="r", marker="o", markersize=10);

plt.subplot(1, 2, 2)
plt.title("Y = X^3", size=20)
plt.xlabel("X", size=20)
plt.ylabel("Y", size=20)
plt.plot(x, y_2, color="b", marker="v", markersize=10);
```

## 6. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง

```python
x = np.arange(-100, 100, 5)
y = np.arange(-200, 200, 10)
```

```python
plt.figure(figsize=[20, 9])

plt.title("Y = X^2 & X = Y^2", size=20)
# plt.xlabel("X", size=20)
# plt.ylabel("Y", size=20)
plt.grid()
plt.plot(y, x**2, color="r", marker="o", markersize=10, label="Y = X^2");
plt.plot(x, y**2, color="b", marker="v", markersize=10, label="X = Y^2");
plt.legend(loc=1, fontsize="xx-large");
```

## 7. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม 

```python
y = np.arange(-200, 200, 10)
x = y**2 + 4*y
```

```python
plt.figure(figsize=[20, 9])
plt.title("X = Y^2 + 4*Y", size=20)
plt.xlabel("X", size=20)
plt.ylabel("Y", size=20)
plt.plot(x, y, color="#bbbb00", linewidth=6);
```

## 8. สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend 

```python
x = np.arange(-100, 100, 5)
y = np.arange(-200, 200, 10)
```

```python
fig = plt.figure(figsize=[20, 9])

plt.title("Y = X^2 & X = Y^2", size=20)
ax1 = fig.add_axes([0.1, 0.1, 0.95, 0.85])
ax1.plot(x, y**3, color="b", marker="v", markersize=10, label="X = Y^3")
ax2 = fig.add_axes([0.15, 0.6, 0.4, 0.3])
ax2.plot(x, x**2+2, color="r", marker="o", markersize=10, label="Y = X^2 + 2")
plt.legend(loc=1, fontsize="xx-large");
```
