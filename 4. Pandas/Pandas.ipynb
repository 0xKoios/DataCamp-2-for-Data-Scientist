## 8. หาจำนวนผู้หญิงและผู้ชายที่รอดชีวิต

```python
df.info()
```

```python
female = df[(df["Sex"] == "female") & (df["Survived"] == 1)]
len(female)
```

```python
male = df[(df["Sex"] == "male") & (df["Survived"] == 1)]
len(male)
```

## 9. หา % การรอดชีวิตของผู้หญิงและผู้ชาย

```python
round((len(female)/len(df["Sex"] == "female"))*100, 2)
```

```python
round((len(male)/len(df["Sex"] == "male"))*100, 2)
```
