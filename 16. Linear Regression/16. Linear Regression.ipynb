12.สร้าง train/test split ของบ้าน สามารถลองทดสอบ 70:30, 80:20, 90:10 ratio ได้ตามใจชอบ

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
columns
```

```python
X = df["Avg. Area Income"]
y = df["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

13.เทรนโมเดลแบบ Simple Linear Regression (สามารถเลือกคู่ได้ หรือ จะลองทุกคู่ก็ได้)

```python
lm = LinearRegression()
lm.fit(X_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1))
```

14.ทดสอบโมเดลวัดค่า MAE, MSE, RMSE


```python
predict = lm.predict(X_test.values.reshape(-1, 1))
```

```python
print(f"MAE:{metrics.mean_absolute_error(y_test, predict):.2f}")
print(f"MSE:{metrics.mean_squared_error(y_test, predict):.2f}")
print(f"RMSE:{metrics.r2_score(y_test, predict):.2f}")
```

```python
y_test.iloc[0], predict[0]
```

```python
for i in range(10):
    print(f"Predict:{predict[i][0]:.2f} \nLabel:{y_test.iloc[i]:.2f}")
    print(f"Diff:{(abs(predict[i][0]-y_test.iloc[i])/y_test.iloc[i])*100:.2f}%")
    print()
```

15.สร้าง distribution plot เพื่อดูว่า predicted results เป็น normal ไหม

```python
fig = plt.figure(figsize=(10, 5))
sns.distplot(predict, bins=50);
```

16.สร้าง scatter plot และ prediction line ของ simple linear regression

```python
fig = plt.figure(figsize=(10, 10))
plt.scatter(X_test, y_test, color='green', label='data')
plt.plot(X_test, predict, color='red', label='Predicted Regression Line')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.legend();
```
